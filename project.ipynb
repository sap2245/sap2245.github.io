```python
import pandas as pd
```

```python
df = pd.read_csv('Electricity Generation.csv')

# Calculate statistics
mean_value = df['Electricity Generation'].mean()
median_value = df['Electricity Generation'].median()
mode_value = df['Electricity Generation'].mode()

# Print the results
print(f"Mean: {mean_value:.2f}")
print(f"Median: {median_value:.2f}")
print(f"Mode: {mode_value.iloc[0]:.2f}")
```

```python
import csv
from statistics import mean, median
from collections import Counter

# Read CSV file and data
data = []
with open('Electricity Generation.csv', 'r', encoding='utf-8-sig') as file:
    csv_reader = csv.reader(file)
    next(csv_reader)  
    for row in csv_reader:
        if row: #check row has variable 
            data.append(float(row[0]))

# Calculate statistics
mean_value = mean(data)
median_value = median(data)

# Calculate mode using Counter function
counter = Counter(data)
mode_value = counter.most_common(1)[0][0] 

# Print the results
print(f"Mean: {mean_value:.2f}")
print(f"Median: {median_value:.2f}")
print(f"Mode: {mode_value:.2f}")
```

```python
import csv
from statistics import mean, median
from collections import Counter
import math

# Read CSV file and data
data = []
with open('Electricity Generation.csv', 'r', encoding='utf-8-sig') as file:
    csv_reader = csv.reader(file)
    next(csv_reader)  
    for row in csv_reader:
        if row:  # check row has variable
            data.append(float(row[0]))

# Calculate statistics
mean_value = mean(data)
median_value = median(data)

# Calculate mode using Counter function
counter = Counter(data)
mode_value = counter.most_common(1)[0][0]

# Calculate number of asterisks (divided by 100 and rounded up)
mean_asterisks = math.ceil(mean_value / 100)
median_asterisks = math.ceil(median_value / 100)
mode_asterisks = math.ceil(mode_value / 100)

# Print the results with asterisks
print(f"Mean: {'*' * mean_asterisks}")
print(f"Median: {'*' * median_asterisks}")
print(f"Mode: {'*' * mode_asterisks}")
```
