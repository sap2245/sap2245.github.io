```python
import pandas as pd
import plotly.express as px
```

```python
happy_df = pd.read_excel("WHR25_Data_Figure_2.1v3.xlsx")
happy_df.head()
```

Dataset for bolivia

```python
happy_df_bolivia = happy_df[happy_df['Country name'] == 'Bolivia']
happy_df_bolivia
```

```python
happy_df_bolivia["Year"].dtype
```

```python
happy_df_bolivia= happy_df_bolivia.sort_values(by='Year', ascending=True)
happy_df_bolivia
```

```python
fig_happy = px.line(happy_df_bolivia, 
                    x='Year', 
                    y='Life evaluation (3-year average)', 
                    title='Happiness over time in Bolivia')

fig_happy.show()
```

Electricity data for bolivia

```python
electricity_acces = pd.read_csv("WB_data.csv")
electricity_acces.head()
```

```python
electricity_by_country_by_year = (
    electricity_acces.drop(columns=["Series Name", "Series Code"])
    .melt(
        id_vars=["Country Name", "Country Code"],
        var_name="Year",
        value_name="Access to electricity (% of population)",
    )
    .dropna()
)

electricity_by_country_by_year
```

```python
electricity_by_country_by_year["Year"] = electricity_by_country_by_year["Year"].str[:4].astype(int)
electricity_by_country_by_year
```

```python
electricity_acces_bolivia= electricity_by_country_by_year[electricity_by_country_by_year['Country Name'] == 'Bolivia']
electricity_acces_bolivia
```

```python
electricity_acces_bolivia= electricity_acces_bolivia.sort_values(by='Year', ascending=True)
electricity_acces_bolivia
```

```python
fig_electricty_bolivia = px.line(electricity_acces_bolivia,
                                 x="Year",
                                 y="Access to electricity (% of population)",
                                 title="Access to electricity in Bolivia over time")
fig_electricty_bolivia.show()
```

The two together now

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

country_name = "Bolivia"

# Create figure with secondary y-axis
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Add traces
fig.add_trace(
    go.Scatter(
        x=electricity_acces_bolivia["Year"],
        y=electricity_acces_bolivia["Access to electricity (% of population)"],
        name="Access to electricity",
    ),
    secondary_y=False,
)

fig.add_trace(
    go.Scatter(
        x=happy_df_bolivia["Year"],
        y=happy_df_bolivia["Life evaluation (3-year average)"],
        name="Life evaluation (3-year average)",
    ),
    secondary_y=True,
)

# Add figure title
fig.update_layout(title_text=f"Access to electricity vs. Happiness, {country_name}")

# Set x-axis title
fig.update_xaxes(title_text="Year")

# Set y-axes titles
fig.update_yaxes(title_text="% of population", secondary_y=False)
fig.update_yaxes(title_text="Happiness score", secondary_y=True)

fig.show()
```
